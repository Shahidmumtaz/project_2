```python
# dependencies and setup
import pandas as pd
import csv
import json
from sqlalchemy import create_engine
from config import api_key
import numpy as np
import sqlite3

```

```python
# load csv file into dataframe
csv_file1 = "Resources/unemp_data.csv"
unemp_df = pd.read_csv(csv_file1, encoding="utf8")
unemp_df.head()
```

```python
unemp_cols = unemp_df[["year", "value"]].round(decimals=2)
unemployment= unemp_cols.rename(columns={"value": "unemp_rate"})
unemployment.to_csv(r"C:\Users\faril\Desktop\project 2 draft\unemployment.csv", encoding="utf-8")
unemployment
```

```python
# load csv file into dataframe
csv_file2 = "Resources/Mortgage_interest_byincome.csv"
mort_int_df = pd.read_csv(csv_file2, encoding="utf8")
mort_int_df
```

```python
mort_int_cols = mort_int_df[["year", "value"]].round(decimals=2)
mortgage_interest = mort_int_cols.rename(columns={"value": "mortgage_interest"})
mortgage_interest.to_csv(r"C:\Users\faril\Desktop\project 2 draft\mortgage_interest.csv", encoding="utf-8")

mortgage_interest
```

```python
# load json file into dataframe
# used API key to parse data from https://apps.bea.gov/api/data
#url = "https://apps.bea.gov/api/data/?UserID=DADBF8F4-419D-4280-8BD8-2E7C383ED68F&method=GetData&datasetname=Regional&TableName=SAGDP2N&LineCode=1&Year=LAST5&GeoFips=STATE&ResultFormat=json"
# json_file = "Resources/GDP_per_state.json"
# GDP_state = pd.read_json(json_file)
# GDP_state_df = open("Resources/GDP_states.csv", "w", encoding="utf8")
# csvwriter = csv.writer(GDP_state_df)

csv_file3 = "Resources/GDP_per_state.csv"
GDP_states_df = pd.read_csv(csv_file3, encoding="utf8")
GDP_cols = GDP_states_df[["GeoName", "Year", "DataValue"]]
GDP_states_cleaned = GDP_cols.rename(columns={"GeoName": "state", "Year":"year", "DataValue":"GDP"})
GDP_states_cleaned
```

```python
GDP_states_grouped = GDP_states_cleaned.groupby(["state", "year", "GDP"]).sum()
GDP_states_grouped
```

```python
# import geo data

csv_file4 = "Resources/lon_lat.csv"
Geo_coords = pd.read_csv(csv_file4, encoding="utf8")
Geo_coords.to_csv(r"C:\Users\faril\Desktop\project 2 draft\Geo_coords.csv", encoding="utf-8", index=False)
Geo_coords.round(decimals=2)
```

```python
lat_lon = Geo_coords[["state", "latitude", "longitude"]]
lat_lon.to_csv(r"C:\Users\faril\Desktop\project 2 draft\lat_lon.csv", encoding="utf-8")

lat_lon.round(decimals=2)
```

```python
state_pop = Geo_coords[["state", "pop_1990", "pop_2000", "pop_2010", "pop_2018"]].round(-6)
state_pop_in_m = state_pop.rename(columns={"pop_1990": "pop_1990_m", "pop_2000":"pop_2000_m", "pop_2010":"pop_2010_m", "pop_2018":"pop_2018_m" })
state_pop_in_m.to_csv(r"C:\Users\faril\Desktop\project 2 draft\state_pop_in_m.csv", encoding="utf-8")

state_pop_in_m
```

```python
state_pop_30m = state_pop_in_m[(state_pop_in_m["pop_1990_m"]==3000000) & (state_pop_in_m["pop_2000_m"]==3000000) & (state_pop_in_m["pop_2010_m"]==3000000) & (state_pop_in_m["pop_2018_m"]==3000000)]
state_pop_30m.to_csv(r"C:\Users\faril\Desktop\project 2 draft\state_pop_30m.csv", encoding="utf-8")

state_pop_30m
```

```python
GDP_states = pd.merge(GDP_states_cleaned, Geo_coords, on="state", how="inner")
GDP_states.to_csv(r"C:\Users\faril\Desktop\project 2 draft\GDP_states.csv", encoding="utf-8")

GDP_states
```

```python
GDP_states_grouped =GDP_states.groupby(["state", "year", "GDP"]).sum()
GDP_states_grouped.round(decimals=2)
```

```python
data_merged = pd.merge(unemployment, mortgage_interest, on=["year"], how="outer")
data_merged_final = data_merged.rename(columns={"value_x": "unemployement", "value_y": "mortgage_interest"})
data_merged_final
```

```python

#final_data =GDP_states.join(data_merged_final, sort=False)

final_data = pd.merge(GDP_states, data_merged_final, on="year", how="inner")
final_data.to_csv(r"C:\Users\faril\Desktop\project 2 draft\final_data.csv", encoding="utf-8")

final_data.round(decimals=2)
```
