```python
# dependencies and setup
import pandas as pd
import csv
import json
from sqlalchemy import create_engine
from config import api_key

```

```python
# load csv file into dataframe
csv_file1 = "Resources/unemp_data.csv"
unemp_df = pd.read_csv(csv_file1, encoding="utf8")
unemp_df.head()
```

```python
unemp_cols = unemp_df[["Series ID", "year", "value"]]
unemp_r = unemp_cols.groupby("year").mean()
unemployment= unemp_r.rename(columns={"value": "unemp_rate"})
unemployment.round(decimals=2)
```

```python
# load csv file into dataframe
csv_file2 = "Resources/Mortgage_interest_byincome.csv"
mort_int_df = pd.read_csv(csv_file2, encoding="utf8")
mort_int_df
```

```python
mort_int_cols = mort_int_df[["Series ID", "year", "value"]]
mortgage_int = mort_int_cols.groupby("year").mean()
mortgage_int.dropna(how="all")
mortgage_interest = mortgage_int.rename(columns={"value": "mortgage_interest"})
mortgage_interest.round(decimals=2)
```

```python
GDP_states = GDP_states_cleaned.groupby(["state", "year", "GDP"]).sum()
GDP_states.round(decimals=2)
```

```python
# import geo data

csv_file4 = "Resources/lon_lat.csv"
Geo_coords = pd.read_csv(csv_file4, encoding="utf8")
Geo_coords.round(decimals=2)
```
